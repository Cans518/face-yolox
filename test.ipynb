```python
import requests
import base64
import cv2
import json
```

```python
img = cv2.imread('test1.jpg')
img_encode = cv2.imencode('.jpg', img)[1]
data_encode = base64.b64encode(img_encode)
data_encode = data_encode.decode('utf-8')
# data_encode = 'data:image/jpeg;base64,' + data_encode
d = {'image': data_encode}
headers = {'Content-Type': 'application/json'}
r = requests.post('http://demo.mzee.top:8080/api/v1/detect', data=json.dumps(d), headers=headers)
print(r.text)
```

```python
img_out = base64.b64decode(r.json()['image'])
# Display the OpenCV image using inbuilt methods.
import numpy as np
img_out = cv2.imdecode(np.frombuffer(img_out, dtype=np.uint8), -1)
cv2.imwrite('test1_out.jpg', img_out)
```
